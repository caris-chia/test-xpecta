```python
import pandas as pd
from geopy.geocoders import GoogleV3

# Cargar el archivo de datos
file_path = '../output/offers.csv'
apartments_data = pd.read_csv(file_path)

# Configurar el geolocalizador con Google Maps
geolocator = GoogleV3(api_key='API_KEY')

def get_neighborhood_google(lat, lng):
    try:
        location = geolocator.reverse((lat, lng), exactly_one=True)
        address = location.raw['address_components']
        # Extraer el barrio si existe en los componentes de la dirección
        for component in address:
            if 'sublocality_level_1' in component['types']:
                return component['long_name']
        return None
    except Exception as e:
        print(f"Error: {e}")
        return None

apartments_data['neighborhood'] = apartments_data.apply(
    lambda row: get_neighborhood_google(row['coordinates_lat'], row['coordinates_lng']), axis=1
)
```

```python
import pandas as pd
from geopy.geocoders import GoogleV3

# Cargar el archivo de datos
file_path = '../prueba_tecnica_xpecta/ofertas-1731276673.798788.csv'
apartments_data = pd.read_csv(file_path)

# Configurar el geolocalizador con Google Maps
geolocator = GoogleV3(api_key='API_KEY')

def get_neighborhood_google(lat, lng):
    # Verificar si las coordenadas son válidas
    if pd.isna(lat) or pd.isna(lng) or (lat == 0.0 and lng == 0.0):
        return None  # Devuelve None si las coordenadas no son válidas

    try:
        location = geolocator.reverse((lat, lng), exactly_one=True)
        address = location.raw['address_components']
        
        # Intentar extraer el barrio
        for component in address:
            if 'neighborhood' in component['types']:
                return component['long_name']
            elif 'sublocality_level_1' in component['types']:
                neighborhood = component['long_name']
            elif 'sublocality_level_2' in component['types']:
                return component['long_name']
        
        # Si no se encuentra el barrio específico, devolver lo que esté en sublocality_level_1
        return neighborhood
    except Exception as e:
        print(f"Error: {e}")
        return None

# Aplicar la función al DataFrame
apartments_data['neighborhood'] = apartments_data.apply(
    lambda row: get_neighborhood_google(row['coordinates_lat'], row['coordinates_lng']), axis=1
)
```
